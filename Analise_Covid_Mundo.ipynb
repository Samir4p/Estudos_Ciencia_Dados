```python
import pandas as pd
import numpy as np
```

# Analise de Numeros de Casos confirmados no Mundo

#Dados = "https://github.com/CSSEGISandData/COVID-19"

```python
casos = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
casos.head(31)
```

```python
casos.tail()
```

```python
# Vamos verificar o tamanho do dataset, depois iremos usar essa função para 
# manter atualizado os dados, pois aumenta a cada dia as culunas

casos.shape
```

```python
# Vamos utilizar a função "iloc" para acessar os dados por linhas e coulnas
# Utilizando o numpy vamos somar as colunas
casos = np.sum(casos.iloc[ : , 4 : casos.shape[1]])

# Vamos transformar as datas em index
casos.index = pd.DatetimeIndex(casos.index)

#Vamos mostrar as ultimas linhas do dataset
casos.tail(10)
```

```python
# Vamos plotar os dados para analise
casos.plot()
```

# Verificando os obitos

```python
# Analisando os dados sobre obitos abrindo dataset
mortes = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
#Analisando as ultimas linhas 
mortes.tail()
```

```python
#Analisando o tamanho do dataset
mortes.shape
```

```python
# Vamos utilizar a função "iloc" para acessar os dados por linhas e coulnas
# Utilizando o numpy vamos somar as colunas
mortes = np.sum(mortes.iloc[ : , 4 : mortes.shape[1]])

# Vamos transformar as datas em index
mortes.index = pd.DatetimeIndex(casos.index)

#Vamos mostrar as ultimas linhas do dataset
mortes.tail(10)
```

```python
mortes.plot()
```

```python
# Vamos cruzar as informações sobre os dois dataFrame para uma analise mais profunda

mortalidade = (mortes/casos)*100
```

```python
mortalidade.plot()
```

# Analisaremos os dados dos recuperados

```python
#Analisando os recuperados
recuperados = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
#Tamanho do dataset
recuperados.shape
```

```python
# Vamos utilizar a função "iloc" para acessar os dados por linhas e coulnas
# Utilizando o numpy vamos somar as colunas
recuperados = np.sum(recuperados.iloc[ : , 4 : recuperados.shape[1]])

# Vamos transformar as datas em index
recuperados.index = pd.DatetimeIndex(recuperados.index)

#Vamos mostrar as ultimas linhas do dataset
recuperados.tail(10)
```

```python
#Plotando os dataset
recuperados.plot()
```

# Analisaremos os ativos para termos uma real visão dos casos curados

```python
#Analisndo os ativos
ativos = casos - (recuperados+mortes)
```

```python
#Ultimas linhas do dataset
ativos.tail()
```

```python
#Plotando as informações
ativos.plot()
```

```python
# Vamos usar a função "concat" e juntar todos os dataframes construidos em um só

df = pd.concat([casos, ativos, recuperados, mortes], axis=1)

# Colocar os nomes das colunas
df.columns = (['casos', 'ativos', 'recuperados',' mortes'])

# Vamos analisar as ultimas linhas do dataset
df.tail()
```

```python
# Plotar as informações geradas

df.plot()
```
